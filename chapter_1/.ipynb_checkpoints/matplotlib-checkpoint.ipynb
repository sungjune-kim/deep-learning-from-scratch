```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10,0.1)
y1 = np.cos(x)
y2 = np.sin(x)
```

```python
plt.plot(x,y1, label="cos")
plt.plot(x,y2, linestyle="--",label="sin")
plt.xlabel("x")
plt.ylabel("y")
plt.title("cos & sin")
plt.legend()
plt.show()
```
