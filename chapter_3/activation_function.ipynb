```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    return np.array(x>0, dtype=np.int)
```

```python
x = np.arange(-5,5,0.1)
y1 = step_function(x)
```

```python
#plt.plot(x,y1)
#plt.ylim(-0.1,1.1)
#plt.title("Step Function")
#plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
y2 = sigmoid(x)
```

```python
#plt.plot(x,y2)
#plt.title("Sigmoid")
#plt.show()
```

----------------------------------------------

#### Step Funciton vs Sigmoid
##### 공통점
<ul> 입출력이 0과 1 사이 </ul>
<ul> 입력이 작으면 0에, 크면 1에 가까운 출력 </ul>
<ul> 비선형 함수 (non-linear) </ul>

##### 차이점 
<ul> sigmoid는 연속적인 실수를 돌려준다 </ul>

### 활성함수로 비선형 함수를 사용해야 하는 이유
선형 함수의 문제는 층을 아무리 깊게 해도 '은닉층이 없는 네트워크'로도 똑같은 기능을 한다. 
h(x) = cx 를 활성화함수로 했을때 3층 네트워크는 y(x) = h(h(h(x))) = c * c * c * x 즉 선형 식인 y(x)=ax와 똑같은 식이다.

# ReLU (Rectified Linear Unit) 

```python
def relu(x):
    return np.maximum(0,x)
```

```python
y3 = relu(x)
```
